```python
import random

fh = open('Vocabulary_list.csv')
wd_list = fh.readlines()
wd_list.pop(0)
wd_set = set(wd_list)
fh = open('Vocabulary_set.csv','w')
fh.writelines(wd_set)


def create_def(s):
    word , defination = s.split(',', 1)
    return word , defination.rstrip()

def get_def_pop(word_list, word_dic):
    random_index = random.randrange(len(word_list))
    word = word_list[random_index]
    word_list.remove(word)
    defination = word_dic.get(word)
    return word , defination


word_dic = {}


for i in wd_set:
    word , defination = create_def(i)
    word_dic[word] = defination

NOQ = 0
SCORE = 0
    
while True:
    word_list = list(word_dic)
    choice_list = []
    for i in range(4):
        word , defination = get_def_pop(word_list, word_dic)
        choice_list.append(defination)
        
    random.shuffle(choice_list)
    
    print(word)
    print('----------------------------------')
    
    for index , choice in enumerate(choice_list):
        print(index+1 , choice)
        
    choice = int(input('\nchoose your options 1 , 2 , 3 , 4 or 0 to exit'))
    
   
    if choice == 0:
        print('RESULTS........................')
        print('Total number of questions ', NOQ)
        print('Total correct ', SCORE)
        try:
            if SCORE/NOQ >= 0.6:
                print('Pass')
            else:
                print('Fail')
        except:
            print('you are so stupid')
        break
        
    elif choice_list[choice -1] == defination:
        print('correct')
        NOQ = NOQ + 1
        SCORE = SCORE + 1
        print('=============================')
    else:
        print('incorrect')
        NOQ = NOQ + 1
        print('=============================')
        
    



    




```
